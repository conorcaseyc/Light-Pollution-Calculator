# Scatter Plot of Population Density (people per $km^2$) - Photopollution ($\frac{mags} {arcsec^2}$), and Population (People in the Area) - Photopollution (Normalization of Walker's Law) 

This notebook will generate two different scatter plots using photopollution data collected with our handheld Sky Quality Metre (SQM), one against population density and the other against the normalization of Walker's Law, throughout the provience of Munster.

```python
#Import Dependencies 
from SQM import *
import matplotlib.pyplot as plt
```

## Scatter Plot of Population Density (people per $km^2$) - Photopollution ($\frac{mags} {arcsec^2}$)

```python
plt.scatter(population_density, sqm)
plt.xlabel("Population Density (people per km^2)")
plt.ylabel("Photopollution (mags / arcsec^2)")
plt.plot(unique(population_density), poly1d(p1)(unique(population_density)))
plt.errorbar(population_density, sqm, stddev, color='g', linestyle='None')
plt.text(0,14.5, "r = " + str(round(correlation, 3)))
```

## Population (people in the area) - Photopollution ($\frac{mags} {arcsec^2}$) (Normalization of Walker's Law)

```python
plt.scatter(population, sqm)
plt.xlabel("Population (People in the Area)")
plt.ylabel("Photopollution (mags / arcsec^2)")
plt.plot(unique(population), poly1d(p2)(unique(population)))
plt.errorbar(population, sqm, stddev, color='g', linestyle='None')
plt.text(0,14.5, "r = " + str(round(correlation2, 3)))
```
